```python
import numpy as np
import matplotlib.pyplot as plt

# Dimensions
L = 10    # longueur du vaisseau
H = 2     # hauteur
x = np.linspace(0, L, 500)
y = np.linspace(-H/2, H/2, 100)

# Géométrie de l'anévrisme
def aneurysm_profile(x, height=0.5, width=1.0, center=5.0):
    return height * np.exp(-((x - center)**2) / (2 * width**2))

X, Y = np.meshgrid(x, y)
vessel_upper = H/2 + aneurysm_profile(x)

plt.figure(figsize=(12, 3))
plt.fill_between(x, -H/2, vessel_upper, color='lightblue', alpha=0.7)
plt.plot(x, vessel_upper, 'b')
plt.plot(x, -H/2 * np.ones_like(x), 'b')
plt.title("Modèle 2D simplifié d’un vaisseau avec anévrisme")
plt.xlabel("Longueur (cm)")
plt.ylabel("Hauteur (cm)")
plt.grid(True)
plt.show()
```
